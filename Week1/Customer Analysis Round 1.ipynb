```python
import pandas as pd
import numpy as np
#Read the three files into python as dataframes
file1=pd.read_csv('file1.csv')
file2=pd.read_csv('file2.csv')
file3=pd.read_csv('file3.csv')

#Show the DataFrame's shape.
display(file1.shape)
display(file2.shape)
display(file3.shape)

#Standardize header names.
head1=[]
head2=[]
head3=[]
for i in range(len(file1.columns)):
    head1.append(file1.columns[i].lower())
    head2.append(file2.columns[i].lower())
    head3.append(file3.columns[i].lower())
file1.columns=head1
file2.columns=head2
file3.columns=head3

file1.columns = file1.columns.str.replace(' ','_')
file2.columns = file2.columns.str.replace(' ','_')
file3.columns = file3.columns.str.replace(' ','_')

#Rearrange the columns in the dataframe as needed
file2=file2[['customer','st', 'gender', 'education', 'customer_lifetime_value',
            'income', 'monthly_premium_auto', 'number_of_open_complaints',
            'policy_type', 'vehicle_class', 'total_claim_amount']]
file3=file3[['customer','state', 'gender', 'education', 'customer_lifetime_value',
            'income', 'monthly_premium_auto', 'number_of_open_complaints',
            'policy_type', 'vehicle_class', 'total_claim_amount']]
file2 = file2.rename(columns={'st':'state'})
file1 = file1.rename(columns={'st':'state'})

print(file2.columns==file1.columns)
print(file3.columns==file1.columns)

#Concatenate the three dataframes

data = pd.concat([file1, file2,file3], axis=0)
display(data)

#Which columns are numerical?

Data_Type = data.select_dtypes(np.number)
display(Data_Type.columns)

#Which columns are categorical?
Data_Type2 = data.select_dtypes('object')
display(Data_Type2.columns)

#Delete the column education and the number of open complaints from the dataframe.

data = data.drop(['education','number_of_open_complaints'], axis=1)

#Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.

data['customer_lifetime_value'] =  pd.to_numeric(data['customer_lifetime_value'], errors='coerce')
data['customer_lifetime_value'] = round((data['customer_lifetime_value'] * 100),2)

#Check for duplicate rows in the data and remove if any.

display(data.duplicated().sum())
data= data.drop_duplicates() 

#Filter out the data for customers who have an income of 0 or less.

mo= data[data['income']<=0]
display(mo)

```
